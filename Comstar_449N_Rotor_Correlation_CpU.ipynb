```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Step 1: Read data from the Excel file
file_location = "/content/drive/MyDrive/Colab Notebooks/Quality_Management_Systems/Products/Comstar_449N_Correlation_Cpk_LowerBound(Cpu)_Height,Perpendicularity/Input/Comstar_449N_Data.xlsx"
df = pd.read_excel(file_location)

# Step 2: Calculate the required statistical measures for Stack_Height and Perpendicularity
stack_height_data_cavity_a = df["COM_449N_30_ROT_S_SHT_CA_DMIC0"]
stack_height_data_cavity_h = df["COM_449N_30_ROT_S_SHT_CH_DVCL0"]
perpendicularity_data_cavity_a = df["COM_449N_30_ROT_S_PTY_CA_DIGA0"]
perpendicularity_data_cavity_h = df["COM_449N_30_ROT_S_PTY_CH_DSYH0"]

# Rest of the code for histograms as before...

# Step 6: Create a 4x4 column correlation matrix
column_names = ["COM_449N_30_ROT_S_SHT_CH_DVCL0", "COM_449N_30_ROT_S_PTY_CH_DSYH0", "COM_449N_30_ROT_S_SHT_CA_DMIC0", "COM_449N_30_ROT_S_PTY_CA_DIGA0"]
correlation_matrix = df[column_names].corr()

# Define the custom colormap with red, yellow, and blue
colors = [(0, 'red'), (0.5, 'yellow'), (1, 'blue')]
cmap = LinearSegmentedColormap.from_list('CustomCmap', colors)

# Step 7: Plot the correlation matrix with the custom colormap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap=cmap, vmin=-1, vmax=1)

# Add color bar
cbar = plt.colorbar()
cbar.set_label('Correlation', rotation=270, labelpad=15)

# Set column and row labels
plt.xticks(np.arange(len(column_names)), column_names, rotation=45, ha='right')
plt.yticks(np.arange(len(column_names)), column_names)

# Add correlation values to the plot
for i in range(len(column_names)):
    for j in range(len(column_names)):
        plt.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='black')

# Add title and show the plot
plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("/content/drive/MyDrive/Colab Notebooks/Quality_Management_Systems/Products/Comstar_449N_Correlation_Cpk_LowerBound(Cpu)_Height,Perpendicularity/Output/Main_Output/Correlation_Matrix.png", format='png')
plt.show()
```
